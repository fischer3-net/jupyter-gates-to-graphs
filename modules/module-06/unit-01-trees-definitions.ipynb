## 1 · What Is a Tree? Five Equivalent Definitions

A **tree** is simultaneously defined by five equivalent conditions. Any one of
them implies all the others — they are five faces of the same object.

Let $G = (V, E)$ be a graph with $n = |V|$ vertices. The following are equivalent:

| # | Statement |
|:---:|---|
| 1 | $G$ is **connected and acyclic** |
| 2 | $G$ is **connected** and $\|E\| = n - 1$ |
| 3 | $G$ is **acyclic** and $\|E\| = n - 1$ |
| 4 | There is **exactly one path** between every pair of vertices |
| 5 | $G$ is **connected**, but removing any single edge disconnects it |

Condition 5 is particularly useful for security reasoning: every edge in a tree
is a **bridge** — a critical connection whose removal severs the graph. In a tree
network topology, every link is a single point of failure.

### The Fundamental Equation

$$|E| = |V| - 1$$

A tree on $n$ vertices has exactly $n - 1$ edges. This is provable by induction
(which we will see in the extension challenge), and it has a useful corollary:

$$|E| < |V| - 1 \Rightarrow G \text{ is disconnected (a forest)}$$
$$|E| > |V| - 1 \Rightarrow G \text{ contains a cycle}$$

This gives a fast sanity check on any graph claiming to be a tree.

### Forests

A **forest** is a graph that is acyclic but not necessarily connected —
a disjoint union of trees. Each connected component of a forest is a tree.

**Security example.** A certificate chain forms a tree rooted at the Certificate
Authority. A PKI infrastructure with multiple CAs forms a forest. Each component
is independently rooted — compromise of one root CA does not affect the others.


## 2 · Rooted Trees and Tree Vocabulary

An **unrooted tree** is just a connected acyclic graph — symmetric, no preferred
direction. A **rooted tree** designates one vertex $r$ as the **root**, which
imposes a natural parent-child hierarchy: every vertex except the root has a
unique parent, and the root has none.

### Structural Terms

For a rooted tree $T$ with root $r$:

- **Root $r$:** the unique vertex at depth 0; has no parent
- **Leaf:** a vertex with no children (degree 1 in the unrooted tree, or depth = $h(T)$)
- **Internal node:** a non-leaf vertex — has at least one child
- **Depth of $v$:** the number of edges on the unique path from $r$ to $v$
- **Height $h(T)$:** the maximum depth of any leaf — the longest root-to-leaf path
- **Level $\ell$:** the set of all vertices at depth $\ell$
- **Subtree $T_v$:** the tree formed by vertex $v$ and all its descendants

### Binary Trees

A **binary tree** is a rooted tree where every vertex has at most 2 children —
conventionally called **left child** and **right child**.

$$\forall v \in V,\; |\text{children}(v)| \leq 2$$

A **full binary tree** has every internal node with exactly 2 children.
A **complete binary tree** fills all levels fully except possibly the last,
which is filled from the left.

### Why Height Matters

In a binary tree with $n$ leaves:
- **Minimum height** (complete/balanced): $h = \lceil \log_2 n \rceil$
- **Maximum height** (completely skewed): $h = n - 1$

The height controls the worst-case cost of any root-to-leaf operation.
A balanced tree gives $O(\log n)$ depth; a degenerate one gives $O(n)$.
This distinction is the difference between efficient and inefficient classifiers,
parsers, and search structures — and between fast and slow security policy lookups.


## 3 · Tree Traversals

A **traversal** visits every vertex of a tree in a systematic order.
For binary trees, three canonical traversals exist — they differ only in
*when* the root is processed relative to its subtrees.

### Pre-order: Root → Left → Right

$$\text{pre-order}(T_v) = [v] + \text{pre-order}(T_{\text{left}}) + \text{pre-order}(T_{\text{right}})$$

Visit the root *before* its children. Pre-order naturally produces a prefix
(Polish) notation of an expression tree.

**Security use:** generating a structured threat model report where the top-level
category is listed before its sub-categories — the structure mirrors the tree.

### In-order: Left → Root → Right

$$\text{in-order}(T_v) = \text{in-order}(T_{\text{left}}) + [v] + \text{in-order}(T_{\text{right}})$$

Visit the root *between* its children. In-order traversal of a binary search tree
yields a sorted sequence.

**Security use:** evaluating an expression tree in natural reading order —
the form a parsed policy expression takes when serialised back to text.

### Post-order: Left → Right → Root

$$\text{post-order}(T_v) = \text{post-order}(T_{\text{left}}) + \text{post-order}(T_{\text{right}}) + [v]$$

Visit the root *after* its children. Post-order naturally matches the evaluation
order of an expression tree — children must be evaluated before their parent.

**Security use:** evaluating a policy expression bottom-up, which is how a Boolean
policy engine computes the final access decision — leaf predicates first,
compound expressions last.

### Level-order (BFS)

Visit vertices level by level, left to right within each level. This is exactly
BFS from the root. Level-order traversal is natural for complete binary trees
stored as arrays.


## 4 · Spanning Trees

Given a connected graph $G = (V, E)$, a **spanning tree** $T = (V, E')$
is a subgraph that:
- Contains **all** vertices of $G$ ($V(T) = V(G)$)
- Is a tree ($T$ is connected and acyclic, $|E'| = |V| - 1$)

A spanning tree is the *minimal connected subgraph* of $G$ — the fewest edges
needed to keep all vertices connected.

### Properties

- Every connected graph has at least one spanning tree (found by DFS or BFS)
- A graph has a unique spanning tree if and only if it is itself a tree
- Removing any edge of a spanning tree disconnects the graph
- A **minimum spanning tree** (MST) minimises $\sum_{e \in E'} w(e)$ over all spanning trees

### Security Relevance

**Minimum spanning tree as secure backbone.** Given a network graph with link costs,
the MST identifies the cheapest set of links that maintains full connectivity.
This is relevant for designing resilient networks — the MST is the backbone
that carries full connectivity with minimum infrastructure.

**Spanning tree protocol (STP).** In Ethernet networks, the Spanning Tree Protocol
is a direct application: it constructs a spanning tree of the network graph to
eliminate cycles (which would cause broadcast storms). STP vulnerabilities arise
when an attacker can influence the root bridge election, causing traffic to route
through a compromised node — a graph-theoretic attack on the spanning tree structure.

**BFS spanning tree = shortest-hop tree.** The BFS tree from a source $s$ is a
spanning tree of the reachable subgraph in which every path from $s$ to any vertex $v$
is a shortest (minimum-hop) path. This is exactly what Module 05 Unit 01 built —
the BFS tree was a spanning tree of the reachability subgraph.

### 1 · Tree Properties Checker

We build three candidate graphs and verify which satisfy the tree conditions.
For the valid tree we compute the full structural profile — depth, height,
leaves, and internal nodes — and verify $|E| = |V| - 1$.


```python
# ── 1 · Tree Properties Checker ──────────────────────────────────────────────

def tree_profile(G, root=None):
    """Compute full structural profile of a tree rooted at root."""
    if not nx.is_tree(G):
        return None
    if root is None:
        root = list(G.nodes())[0]

    T = nx.bfs_tree(G, root)   # directed BFS tree for depth computation
    depths  = nx.single_source_shortest_path_length(G, root)
    height  = max(depths.values())
    leaves  = [v for v in G.nodes() if G.degree(v) == 1 and v != root]
    if G.degree(root) == 0: leaves.append(root)  # single-node tree
    internal = [v for v in G.nodes() if v not in leaves]

    return {
        'n': G.number_of_nodes(), 'm': G.number_of_edges(),
        'root': root, 'height': height,
        'leaves': sorted(leaves), 'internal': sorted(internal),
        'depths': depths,
        'is_tree': True,
        'E_eq_V_minus_1': G.number_of_edges() == G.number_of_nodes() - 1,
    }

# Three candidate graphs
G_tree = nx.Graph()
G_tree.add_edges_from([
    ('A','B'), ('A','C'), ('B','D'), ('B','E'),
    ('C','F'), ('C','G'), ('D','H'),
])

G_cyclic = nx.Graph()
G_cyclic.add_edges_from([
    ('A','B'), ('A','C'), ('B','C'), ('C','D'),
])

G_forest = nx.Graph()
G_forest.add_edges_from([
    ('A','B'), ('A','C'),          # tree 1
    ('D','E'),                     # tree 2 (disconnected)
])

candidates = [
    ('G_tree   (valid tree)',   G_tree),
    ('G_cyclic (has a cycle)',  G_cyclic),
    ('G_forest (disconnected)', G_forest),
]

print(f"{'Graph':<35} {'Connected':>10} {'Acyclic':>8} {'|E|=|V|-1':>10} {'Is tree?':>10}")
print("─" * 76)
for name, G in candidates:
    conn  = nx.is_connected(G)
    acyc  = nx.is_forest(G)
    eq    = G.number_of_edges() == G.number_of_nodes() - 1
    is_t  = nx.is_tree(G)
    print(f"  {name:<33} {'✓' if conn else '✗':>10} {'✓' if acyc else '✗':>8} "
          f"{'✓' if eq else '✗':>10} {'✓' if is_t else '✗':>10}")

# Full profile for the valid tree
print()
prof = tree_profile(G_tree, root='A')
print(f"Tree profile (root = A):")
print(f"  |V| = {prof['n']},  |E| = {prof['m']},  |E| = |V|-1: {prof['E_eq_V_minus_1']} ✓")
print(f"  Height:   {prof['height']}")
print(f"  Leaves:   {prof['leaves']}")
print(f"  Internal: {prof['internal']}")
print(f"  Depths:   {dict(sorted(prof['depths'].items(), key=lambda x: x[1]))}")

# ── Visualise the tree ────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 6))
pos = nx.drawing.nx_agraph.graphviz_layout(G_tree, prog='dot', args='-Grankdir=TB')       if False else nx.spring_layout(G_tree, seed=1, k=3)

# Use a hierarchy layout manually
pos = {'A':(4,4), 'B':(2,3), 'C':(6,3), 'D':(1,2), 'E':(3,2),
       'F':(5,2), 'G':(7,2), 'H':(1,1)}

depths_map = prof['depths']
nc = [TS_RED if v == 'A' else
      TS_GREEN if v in prof['leaves'] else
      TS_BLUE
      for v in G_tree.nodes()]

nx.draw_networkx_edges(G_tree, pos, ax=ax, edge_color=TS_GRAY, width=2)
nx.draw_networkx_nodes(G_tree, pos, ax=ax, node_color=nc, node_size=900, alpha=0.92)
nx.draw_networkx_labels(G_tree, pos, ax=ax, font_color='white',
                        font_size=11, font_weight='bold')

# Depth annotations
for v, (x,y) in pos.items():
    ax.text(x, y - 0.35, f"d={depths_map[v]}", ha='center', fontsize=8, color=TS_GRAY)

# Level lines
for level in range(prof['height'] + 1):
    level_verts = [v for v,d in depths_map.items() if d == level]
    if level_verts:
        y_lev = pos[level_verts[0]][1]
        ax.axhline(y_lev, color='#eeeeee', lw=1, zorder=0)
        ax.text(-0.2, y_lev, f'Level {level}', va='center', fontsize=8, color=TS_GRAY)

legend = [mpatches.Patch(color=TS_RED,   label='Root'),
          mpatches.Patch(color=TS_BLUE,  label='Internal node'),
          mpatches.Patch(color=TS_GREEN, label='Leaf')]
ax.legend(handles=legend, loc='upper right', fontsize=9)
ax.set_title(f'Rooted Tree (root=A)  |  |V|={prof["n"]}, |E|={prof["m"]}, '
             f'height={prof["height"]}', pad=10, fontsize=11)
ax.axis('off'); ax.set_facecolor('white'); fig.patch.set_facecolor('white')
plt.tight_layout()
plt.show()
```

The properties table confirms that only `G_tree` satisfies all three conditions:

The tree diagram shows the three-level hierarchy. Vertices at level 0 (root),

### 2 · Traversal Visualizer — Expression Tree

We build a binary expression tree for the Boolean access policy
$A \wedge (T \vee \neg S)$ and perform all three traversals, labelling each
node with its visit order. This makes the traversal sequences concrete
and connects them to policy evaluation order.


```python
# ── 2 · Traversal Visualizer — Boolean Policy Expression Tree ────────────────
#
# Policy: A ∧ (T ∨ ¬S)
#         ∧
#        / \
#       A   ∨
#          / \
#         T   ¬
#              \
#               S

class BinaryNode:
    def __init__(self, val, left=None, right=None):
        self.val   = val
        self.left  = left
        self.right = right

# Build the expression tree
root_tree = BinaryNode('∧',
    left  = BinaryNode('A'),
    right = BinaryNode('∨',
        left  = BinaryNode('T'),
        right = BinaryNode('¬',
            right = BinaryNode('S')
        )
    )
)

def pre_order(node, result=None):
    if result is None: result = []
    if node:
        result.append(node.val)
        pre_order(node.left, result)
        pre_order(node.right, result)
    return result

def in_order(node, result=None):
    if result is None: result = []
    if node:
        in_order(node.left, result)
        result.append(node.val)
        in_order(node.right, result)
    return result

def post_order(node, result=None):
    if result is None: result = []
    if node:
        post_order(node.left, result)
        post_order(node.right, result)
        result.append(node.val)
    return result

pre  = pre_order(root_tree)
ino  = in_order(root_tree)
post = post_order(root_tree)

print("Expression tree for:  A ∧ (T ∨ ¬S)")
print()
print(f"Pre-order  (Root→Left→Right): {pre}")
print(f"  → Prefix notation: {' '.join(pre)}")
print(f"  → Top-level operator first: '{pre[0]}' tells you the overall structure immediately")
print()
print(f"In-order   (Left→Root→Right): {ino}")
print(f"  → Infix notation (with implied parentheses): {' '.join(ino)}")
print(f"  → Natural reading order of the expression")
print()
print(f"Post-order (Left→Right→Root): {post}")
print(f"  → Postfix (RPN) notation: {' '.join(post)}")
print(f"  → Evaluation order: leaves computed first, operators applied to their results")
print(f"  → This is how a policy engine evaluates: A, T, S, ¬, ∨, ∧")

# ── Annotated tree diagram ────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(14, 6))

traversals = [
    ('Pre-order
(Root first)', pre,  'Top-down: structure before content'),
    ('In-order
(Root middle)', ino,  'Natural reading order'),
    ('Post-order
(Root last)', post, 'Bottom-up: leaves before operators'),
]

tree_pos = {
    '∧_0': (3, 4),   '∧': (3,4),
    'A':   (1, 3),
    '∨':   (5, 3),
    'T':   (4, 2),
    '¬':   (6, 2),
    'S':   (6, 1),
}

# Build networkx tree for drawing
GT = nx.DiGraph()
GT.add_edges_from([
    ('∧','A'), ('∧','∨'), ('∨','T'), ('∨','¬'), ('¬','S')
])
pos_t = {'∧':(3,4), 'A':(1,3), '∨':(5,3), 'T':(4,2), '¬':(6,2), 'S':(6,1)}
node_order = ['∧','A','∨','T','¬','S']

for ax, (title, order, subtitle) in zip(axes, traversals):
    visit_rank = {v: i+1 for i, v in enumerate(order)}
    nc_t = [TS_AMBER if GT.nodes[v].get('type') == 'op' else TS_BLUE
            for v in GT.nodes()]

    # Colour by visit order (early = green, late = amber)
    nc_t = []
    for v in GT.nodes():
        rank = visit_rank.get(v, 0)
        if rank <= 2:   nc_t.append(TS_GREEN)
        elif rank <= 4: nc_t.append(TS_BLUE)
        else:           nc_t.append(TS_AMBER)

    nx.draw_networkx_edges(GT, pos_t, ax=ax, edge_color=TS_GRAY, width=2,
                           arrows=True, arrowsize=16)
    nx.draw_networkx_nodes(GT, pos_t, ax=ax, node_color=nc_t,
                           node_size=900, alpha=0.92)
    nx.draw_networkx_labels(GT, pos_t, ax=ax, font_color='white',
                            font_size=12, font_weight='bold')

    # Visit order numbers
    for v, (x,y) in pos_t.items():
        ax.text(x+0.35, y+0.3, str(visit_rank[v]),
                fontsize=11, fontweight='bold', color=TS_RED)

    ax.set_title(f'{title}
{order}
{subtitle}', fontsize=9, pad=8)
    ax.axis('off'); ax.set_facecolor('white')

plt.suptitle('Three Traversals of the Expression Tree: A ∧ (T ∨ ¬S)
'
             'Red numbers = visit order', fontsize=11, fontweight='bold', y=1.02)
fig.patch.set_facecolor('white')
plt.tight_layout()
plt.show()
```

**What do you see?**

The three traversals visit the same six nodes in completely different orders:

- **Pre-order** visits the root $\wedge$ first — you know the top-level operation
  before seeing any operands. This is how you'd describe the policy structure:
  "it's an AND of A with something."

- **In-order** produces the natural reading: `A ∧ T ∨ ¬ S` — the infix expression
  most humans would write (with implied parentheses giving the correct structure).

- **Post-order** visits the leaves first and the root last: `A T S ¬ ∨ ∧`. This is
  the evaluation order — a policy engine evaluating this expression must first check
  $A$ (admin), then check $T$ (time), then check $S$ (suspended), then apply $\neg$,
  then apply $\vee$, then apply $\wedge$. **Post-order is how every bottom-up
  policy evaluator works.**

The red visit-order numbers make the structural difference between the traversals
impossible to miss. Notice that the leaf node $S$ is visited 3rd in post-order
but 6th in pre-order — the same node, very different position in the sequence.


### 3 · Spanning Trees — BFS vs DFS

We compute both BFS and DFS spanning trees of the enterprise network from
Module 05, compare their structures, and explain which is the natural choice
for security analysis and why.


```python
# ── 3 · Spanning Trees — BFS vs DFS ─────────────────────────────────────────

G_net = nx.Graph()
net_edges = [
    ('internet','web_server'), ('internet','api_server'),
    ('web_server','load_balancer'), ('web_server','api_server'),
    ('api_server','db_primary'), ('api_server','log_server'),
    ('api_server','dev_laptop'), ('load_balancer','db_primary'),
    ('db_primary','db_replica'), ('db_primary','backup_server'),
    ('db_primary','admin_workstation'), ('dev_laptop','admin_workstation'),
    ('dev_laptop','log_server'), ('admin_workstation','domain_controller'),
    ('admin_workstation','log_server'), ('log_server','backup_server'),
]
G_net.add_edges_from(net_edges)
source = 'internet'

# Compute spanning trees
bfs_tree = nx.bfs_tree(G_net, source)
dfs_tree = nx.dfs_tree(G_net, source)

print(f"Network graph:  |V|={G_net.number_of_nodes()}, |E|={G_net.number_of_edges()}")
print(f"BFS span. tree: |V|={bfs_tree.number_of_nodes()}, |E|={bfs_tree.number_of_edges()}")
print(f"DFS span. tree: |V|={dfs_tree.number_of_nodes()}, |E|={dfs_tree.number_of_edges()}")
print(f"|E| = |V|-1 for both: {bfs_tree.number_of_edges() == bfs_tree.number_of_nodes()-1} ✓")

# BFS depths vs DFS depths
bfs_depths = dict(nx.single_source_shortest_path_length(G_net, source))
dfs_par = {c: list(dfs_tree.predecessors(c))[0]
           if list(dfs_tree.predecessors(c)) else None
           for c in dfs_tree.nodes()}

def dfs_depth(v, par, memo={}):
    if v == source: return 0
    p = par[v]
    if p is None: return 0
    return 1 + dfs_depth(p, par, memo)

dfs_depths = {v: dfs_depth(v, dfs_par) for v in dfs_tree.nodes()}

print(f"
{'Vertex':<24} {'BFS depth':>10} {'DFS depth':>10}")
print("─" * 46)
for v in sorted(G_net.nodes()):
    bd = bfs_depths.get(v,'?')
    dd = dfs_depths.get(v,'?')
    flag = '  ← deeper in DFS' if isinstance(bd,int) and isinstance(dd,int) and dd > bd else ''
    print(f"  {v:<22}  {bd:>9}  {dd:>9}{flag}")

# ── Visualise side by side ─────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 7))
pos_net = nx.spring_layout(G_net, seed=42, k=2.5)

for ax, tree, title, color in [
    (axes[0], bfs_tree, 'BFS Spanning Tree
(minimum-hop from internet)', TS_BLUE),
    (axes[1], dfs_tree, 'DFS Spanning Tree
(deep exploration)', TS_AMBER),
]:
    # Background: original network
    nx.draw_networkx_edges(G_net, pos_net, ax=ax, edge_color='#dddddd',
                           width=1, alpha=0.5)
    # Spanning tree edges highlighted
    nx.draw_networkx_edges(tree, pos_net, ax=ax, edge_color=color,
                           width=2.8, alpha=0.9, arrows=True, arrowsize=16,
                           connectionstyle='arc3,rad=0.04')
    # Nodes
    nc = [TS_GREEN if v == source else
          TS_RED   if v == 'domain_controller' else TS_BLUE
          for v in G_net.nodes()]
    nx.draw_networkx_nodes(G_net, pos_net, ax=ax, node_color=nc,
                           node_size=800, alpha=0.92)
    nx.draw_networkx_labels(G_net, pos_net, ax=ax, font_size=7,
                            font_color='white', font_weight='bold')
    ax.set_title(title, fontweight='bold', pad=8)
    ax.axis('off'); ax.set_facecolor('white')

legend = [mpatches.Patch(color=TS_GREEN, label='Source (internet)'),
          mpatches.Patch(color=TS_RED,   label='Target (domain_controller)'),
          mpatches.Patch(color='#dddddd',label='Non-tree edge')]
fig.legend(handles=legend, loc='lower center', ncol=3, fontsize=9,
           bbox_to_anchor=(0.5,-0.02))
fig.patch.set_facecolor('white')
plt.suptitle('BFS vs DFS Spanning Trees of the Enterprise Network',
             fontsize=12, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()

print("\nSecurity interpretation:")
print("  BFS spanning tree = minimum-hop attack tree.")
print("  Every root-to-leaf path is a shortest lateral movement route.")
print("  DFS spanning tree = deep exploration — may take longer paths to reach nodes.")
print("  For attack analysis: BFS is preferred (shows the quickest attack routes).")
print("  For dependency checking: DFS is preferred (finds deep dependency chains).")
```

**What do you see?**

Both spanning trees include all 11 vertices with exactly 10 edges ($|V| - 1$) —
the fundamental tree equation holds for both.

The depth table shows the key difference: BFS depths are always minimal (each
vertex is at its true shortest hop distance from `internet`), while DFS depths
can be larger because DFS may take a winding path to reach a vertex before
backtracking. The `domain_controller` has BFS depth 4 but a larger DFS depth —
DFS discovered it via a longer path.

**For security analysis, BFS spanning trees are the natural choice:** they give
every vertex its minimum exposure distance from the attacker's entry point.
DFS spanning trees are better for finding deep dependency chains or detecting
cycles (Module 04 Unit 02 used exactly this).

## Summary

| Concept | Definition | Security / AI meaning |
|---|---|---|
| **Tree** | Connected acyclic graph | Minimal connected structure; every edge is a bridge |
| **$\|E\| = \|V\| - 1$** | Fundamental tree equation | Fast sanity check; every edge is critical |
| **Leaf** | Vertex with no children | Terminal decision; output neuron; certificate endpoint |
| **Height $h(T)$** | Max root-to-leaf depth | Worst-case cost of any root-to-leaf operation |
| **Pre-order** | Root before children | Top-down structure; threat model outline |
| **In-order** | Root between children | Natural infix reading |
| **Post-order** | Root after children | Bottom-up evaluation; policy engine order |
| **Spanning tree** | Connected subgraph with $\|V\|-1$ edges | Minimal network backbone |
| **BFS spanning tree** | Minimal-hop tree from source | Shortest attack distance tree |
