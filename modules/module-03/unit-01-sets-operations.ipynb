| $\{x \in D \mid P(x)\}$ | Set-builder notation | "all x in D such that P(x)" | Elements of D satisfying predicate P |

| $\|A\|$ | Cardinality | "the size of $A$" | Number of elements in $A$ |

| $A \cap B$ | Intersection | "$A$ intersect $B$" | Only elements in **both** $A$ and $B$ |
| $A \setminus B$ | Difference | "$A$ minus $B$" | Elements of $A$ **not** in $B$ |
| $A^c$ or $\overline{A}$ | Complement | "complement of $A$" | All elements of the universe **not** in $A$ |
| $A \triangle B$ | Symmetric difference | "$A$ symmetric difference $B$" | Elements in $A$ or $B$ but **not both** |
| $A \times B$ | Cartesian product | "$A$ cross $B$" | All ordered pairs $(a, b)$ with $a \in A$, $b \in B$ |

> **The connection to Module 02.** The domain $D$ in $\forall x \in D,\; P(x)$ is a set.
> The predicate $P$ partitions $D$ into two sets: $\{x \in D \mid P(x)\}$ (satisfying)
> and $\{x \in D \mid \neg P(x)\}$ (violating). Set theory is the formal foundation
> of everything we built in Modules 01–02.

## 1 · Defining Sets

A **set** is an unordered collection of distinct elements. "Unordered" means
$\{1, 2, 3\} = \{3, 1, 2\}$. "Distinct" means each element appears at most once —
$\{1, 1, 2\}$ is not a valid set notation; it simplifies to $\{1, 2\}$.

### Three Ways to Define a Set

**Roster notation** — list the elements explicitly:

$$A = \{\mathtt{read},\, \mathtt{write},\, \mathtt{execute}\}$$

**Set-builder notation** — specify a domain and a predicate *(from Module 02)*:

$$B = \{x \in \mathbb{Z} \mid 0 \leq x \leq 255\}$$

Read: *"all integers $x$ such that $x$ is between 0 and 255 inclusive."*
This is the set of valid byte values.

**Predicate notation** — name the set by the property its members satisfy:

$$C = \{u \in \text{Users} \mid \text{admin}(u) \wedge \neg\,\text{suspended}(u)\}$$

This is *the set of active admins* — the extension of the predicate
$\text{admin}(u) \wedge \neg\,\text{suspended}(u)$ from Module 02 · Unit 01,
now written as an explicit set.

### Important Sets in Computing

| Set | Symbol | What it contains |
|---|---|---|
| Natural numbers | $\mathbb{N}$ | $0, 1, 2, 3, \ldots$ |
| Integers | $\mathbb{Z}$ | $\ldots, -2, -1, 0, 1, 2, \ldots$ |
| Bytes | $\mathbb{B}$ | $\{0, 1, \ldots, 255\}$ — valid 8-bit values |
| Binary strings of length $n$ | $\{0,1\}^n$ | All $n$-bit sequences — $2^n$ elements |
| Printable ASCII | $\Sigma_{\text{ASCII}}$ | Characters with code points 32–126 |
| Valid inputs | $\mathcal{I}_{\text{valid}}$ | $\{s \in \Sigma^* \mid \text{validate}(s)\}$ |


## 2 · Membership, Subsets, and Set Equality

### Membership — $\in$

$x \in A$ *(x is in A)* means $x$ is one of the elements of $A$.
It is a predicate — true or false for each specific $x$.

$$\mathtt{read} \in \{\mathtt{read},\, \mathtt{write},\, \mathtt{execute}\} \quad \text{True}$$
$$\mathtt{delete} \notin \{\mathtt{read},\, \mathtt{write},\, \mathtt{execute}\} \quad \text{True}$$

### Subsets — $\subseteq$

$A \subseteq B$ *(A is a subset of B)* means every element of $A$ is also in $B$:

$$A \subseteq B \;\equiv\; \forall x,\; x \in A \Rightarrow x \in B$$

Notice: this is a $\forall$ statement. To disprove $A \subseteq B$, find one
$x \in A$ with $x \notin B$ — a counterexample.

**Security reading:** $\text{perms}(\text{user}) \subseteq \text{perms}(\text{role})$ —
the user's actual permissions are a subset of what their role permits.
This is the least privilege property expressed as a subset relationship.

### Set Equality — $=$

$$A = B \;\equiv\; A \subseteq B \;\wedge\; B \subseteq A$$

To prove two sets are equal, prove containment in both directions — a direct
proof technique from Module 00. This is the standard method for all set equality
proofs.

### The Empty Set — $\emptyset$

The empty set $\emptyset = \{\}$ contains no elements.
It is a subset of every set: $\emptyset \subseteq A$ for all $A$.

**Security reading:** $\text{perms}(u) = \emptyset$ — user $u$ has no permissions.
A freshly created account should start here before any roles are assigned.


## 3 · Set Operations

All six operations build new sets from existing ones. Each has a direct
security interpretation.

### Union — $A \cup B$

$$A \cup B = \{x \mid x \in A \vee x \in B\}$$

*All elements in $A$ or $B$ (or both).* Union is set-level OR — it widens.

**Security:** $\text{perms}(\text{admin}) \cup \text{perms}(\text{owner})$ —
the combined permissions of two roles. Adding a role to a user unions their
permission sets.

### Intersection — $A \cap B$

$$A \cap B = \{x \mid x \in A \wedge x \in B\}$$

*Only elements in both $A$ and $B$.* Intersection is set-level AND — it restricts.

**Security:** $\text{requested\_perms} \cap \text{granted\_perms}$ — what a user
actually gets: only the intersection of what they asked for and what they were given.

### Difference — $A \setminus B$

$$A \setminus B = \{x \mid x \in A \wedge x \notin B\}$$

*Elements in $A$ but not in $B$.* Also written $A - B$.

**Security:** $\text{perms}(\text{admin}) \setminus \text{perms}(\text{user})$ —
the permissions an admin has that a regular user does not. This is the
**privilege gap** — the set of capabilities that make admin accounts high-value
targets.

### Complement — $\overline{A}$

$$\overline{A} = \{x \in \mathcal{U} \mid x \notin A\}$$

*Everything in the universe $\mathcal{U}$ not in $A$.*

**Security:** $\overline{\mathcal{I}_{\text{valid}}}$ — all inputs that *fail*
validation. This is the attack surface of an input validation function.

### Symmetric Difference — $A \triangle B$

$$A \triangle B = (A \setminus B) \cup (B \setminus A) = (A \cup B) \setminus (A \cap B)$$

*Elements in $A$ or $B$ but not both* — set-level XOR.

**Security:** $\text{perms}_t \triangle \text{perms}_{t+1}$ — the permission
**delta** after a policy change. A non-empty symmetric difference means something
changed; a security audit should review every element.

### Cartesian Product — $A \times B$

$$A \times B = \{(a, b) \mid a \in A \;\wedge\; b \in B\}$$

*All ordered pairs.* $|A \times B| = |A| \cdot |B|$.

**Security:** $\text{Users} \times \text{Resources}$ — the full access control
matrix. Each pair $(u, r)$ is either permitted or denied. The Cartesian product
enumerates every possible access decision the system must make.


## 4 · Power Sets and Cardinality

### Cardinality — $|A|$

The **cardinality** of a set is the number of elements it contains.

$$|\{\mathtt{read}, \mathtt{write}, \mathtt{execute}\}| = 3$$
$$|\emptyset| = 0$$
$$|\{0, 1\}^8| = 2^8 = 256 \quad \text{(the byte set)}$$

For infinite sets, cardinality becomes more subtle — a topic for graduate
set theory. In this course we work with finite sets.

### Power Set — $\mathcal{P}(A)$

The **power set** of $A$ is the set of all subsets of $A$, including $\emptyset$
and $A$ itself:

$$\mathcal{P}(A) = \{S \mid S \subseteq A\}$$

$$\mathcal{P}(\{a, b, c\}) = \{\emptyset,\, \{a\},\, \{b\},\, \{c\},\, \{a,b\},\, \{a,c\},\, \{b,c\},\, \{a,b,c\}\}$$

**Key fact:** $|\mathcal{P}(A)| = 2^{|A|}$

If $|A| = n$, the power set has $2^n$ elements — each subset corresponds to a
binary choice (in or out) for each of the $n$ elements.

### Security Relevance — Permission Subsets

If a system has $n$ base permissions, the set of all possible permission
assignments is $\mathcal{P}(\text{Permissions})$ — it has $2^n$ elements.

For $n = 10$ permissions: $2^{10} = 1{,}024$ possible permission sets.
For $n = 30$: $2^{30} > 10^9$ — more than a billion possible configurations.

This exponential growth is why access control systems use **role-based** rather
than **permission-based** assignment — you assign users to a small number of
roles rather than choosing from $2^n$ permission subsets. We will formalise
this connection in Unit 02 using relations.


## 5 · Set Laws

Set operations satisfy algebraic laws analogous to Boolean algebra. This is not
a coincidence — the correspondence is exact, and we will formalise it in Unit 03.

| Law | Set form |
|---|---|
| **Commutative** | $A \cup B = B \cup A$ · $A \cap B = B \cap A$ |
| **Associative** | $(A \cup B) \cup C = A \cup (B \cup C)$ |
| **Distributive** | $A \cap (B \cup C) = (A \cap B) \cup (A \cap C)$ |
| **Identity** | $A \cup \emptyset = A$ · $A \cap \mathcal{U} = A$ |
| **Complement** | $A \cup \overline{A} = \mathcal{U}$ · $A \cap \overline{A} = \emptyset$ |
| **De Morgan** | $\overline{A \cup B} = \overline{A} \cap \overline{B}$ · $\overline{A \cap B} = \overline{A} \cup \overline{B}$ |
| **Double complement** | $\overline{\overline{A}} = A$ |
| **Absorption** | $A \cup (A \cap B) = A$ · $A \cap (A \cup B) = A$ |

**De Morgan's Laws for sets** say exactly what they said for logic in Module 01 —
the complement of a union is the intersection of the complements, and vice versa.
The symbols changed ($\cup / \cap$ instead of $\vee / \wedge$, $\overline{A}$
instead of $\neg P$), but the structure is identical.

## 🔐 Security Bridge &nbsp;·&nbsp; \[AC\] \[AE\]

| Set operation | Security meaning | Concrete example |
|---|---|---|
| $\cup$ (union) | Permission grant, role merge | $\text{perms}(r_1) \cup \text{perms}(r_2)$ when user holds both roles |
| $\cap$ (intersection) | Least common access | Mandatory access control intersection of labels |
| $\setminus$ (difference) | Privilege gap | $\text{perms}(\text{admin}) \setminus \text{perms}(\text{user})$ |
| $\overline{A}$ (complement) | Attack surface / denial set | $\overline{\mathcal{I}_{\text{valid}}}$ — all invalid inputs |
| $\triangle$ (sym. difference) | Policy change delta | Permissions added or removed between two policy versions |
| $\times$ (Cartesian product) | Full access matrix | $\text{Users} \times \text{Resources}$ — all access decisions |
| $\mathcal{P}(A)$ (power set) | All possible permission sets | $2^n$ configurations for $n$ permissions |

**Input validation as set containment.** The security claim "all inputs are
validated" is a subset check:

$$\text{actual\_inputs} \subseteq \mathcal{I}_{\text{valid}}$$

A successful injection attack is a counterexample to this subset claim — an
actual input that is not in the valid set. The fuzz testing goal is to find
elements of $\text{actual\_inputs} \cap \overline{\mathcal{I}_{\text{valid}}}$.

### 1 · Set Operations on Permission Sets

We define two permission sets for two roles in a hypothetical system, then
compute all six operations. Python's built-in `set` type models finite sets exactly.


```python
# ── 1 · Set Operations on Permission Sets ────────────────────────────────────

# Base permission universe
UNIVERSE = {
    'read', 'write', 'execute', 'delete',
    'admin_panel', 'view_logs', 'export_data',
    'manage_users', 'reset_passwords', 'configure_system'
}

# Role permission sets
analyst  = {'read', 'view_logs', 'export_data'}
engineer = {'read', 'write', 'execute', 'view_logs', 'delete'}

# Compute all operations
ops = {
    'A ∪ B  (union)':               analyst | engineer,
    'A ∩ B  (intersection)':        analyst & engineer,
    'A \ B  (difference: A−B)':    analyst - engineer,
    'B \ A  (difference: B−A)':    engineer - analyst,
    'A △ B  (sym. difference)':     analyst ^ engineer,
    'complement(A)':                 UNIVERSE - analyst,
}

# ── Display ────────────────────────────────────────────────────────────────────
print(f"Universe  U  ({len(UNIVERSE)} perms): {sorted(UNIVERSE)}")
print(f"Analyst   A  ({len(analyst)} perms):  {sorted(analyst)}")
print(f"Engineer  B  ({len(engineer)} perms): {sorted(engineer)}")
print()

for label, result in ops.items():
    print(f"  {label:<30} ({len(result):>2} perms): {sorted(result)}")

# ── Venn diagram for union, intersection, and A-B ─────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(13, 4.5))
fig.patch.set_facecolor('white')

op_info = [
    ('A ∪ B  (union)
"analyst OR engineer"',
     analyst, engineer, True,  True,  True),
    ('A ∩ B  (intersection)
"analyst AND engineer"',
     analyst, engineer, False, True,  False),
    ('A \ B  (difference)
"analyst only"',
     analyst, engineer, True,  False, False),
]

def venn2(ax, A, B, title, shade_left, shade_mid, shade_right, la='A', lb='B'):
    ax.set_xlim(-2.2, 2.2)
    ax.set_ylim(-1.7, 2.0)
    ax.axis('off')
    ax.set_facecolor('white')

    alpha = 0.55
    if shade_left:
        c = Circle((-0.55, 0), 1.1, facecolor=TS_BLUE, alpha=alpha, zorder=2)
        ax.add_patch(c)
    if shade_right:
        c = Circle((0.55, 0), 1.1, facecolor=TS_AMBER, alpha=alpha, zorder=2)
        ax.add_patch(c)
    if shade_mid and not (shade_left and shade_right):
        # draw overlap only
        from matplotlib.patches import PathPatch
        import matplotlib.path as mpath
        c = Circle((0.0, 0), 0.6, facecolor=TS_GREEN, alpha=alpha, zorder=3)
        ax.add_patch(c)

    # Outlines
    for cx, color in [(-0.55, TS_BLUE), (0.55, TS_AMBER)]:
        c = Circle((cx, 0), 1.1, facecolor='none', edgecolor=color, lw=2.5, zorder=4)
        ax.add_patch(c)

    # Labels
    ax.text(-1.35, 0, la, ha='center', va='center', fontsize=13,
            fontweight='bold', color=TS_BLUE, zorder=5)
    ax.text(1.35, 0,  lb, ha='center', va='center', fontsize=13,
            fontweight='bold', color=TS_AMBER, zorder=5)

    # Element labels (sample)
    only_A  = sorted(A - B)
    only_B  = sorted(B - A)
    both    = sorted(A & B)

    for i, p in enumerate(only_A[:3]):
        ax.text(-1.1, 0.35 - i*0.35, p, ha='center', fontsize=7,
                color='white' if shade_left else TS_GRAY, zorder=6)
    for i, p in enumerate(only_B[:3]):
        ax.text(1.1, 0.35 - i*0.35, p, ha='center', fontsize=7,
                color='white' if shade_right else TS_GRAY, zorder=6)
    for i, p in enumerate(both[:2]):
        ax.text(0, 0.2 - i*0.35, p, ha='center', fontsize=7,
                color='white' if shade_mid else TS_GRAY, zorder=6)

    ax.set_title(title, fontsize=9, fontweight='bold', pad=8)

for ax, (title, A, B, sl, sm, sr) in zip(axes, op_info):
    venn2(ax, A, B, title, sl, sm, sr, la='Analyst', lb='Engineer')

plt.suptitle('Set Operations on Permission Sets', fontsize=12, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Security summary
priv_gap = engineer - analyst
print(f"\nPrivilege gap (Engineer \ Analyst) = {sorted(priv_gap)}")
print(f"These {len(priv_gap)} permissions are the high-value targets")
print(f"not available to an analyst — what an attacker gains by")
print(f"compromising an engineer account vs an analyst account.")
```

**What do you see?**

- **Union** (left): the full coloured area — every permission either role holds.
  This is what a user with *both* roles gets.
- **Intersection** (centre): only the overlap — `read` and `view_logs`, the
  permissions shared by both roles.
- **Difference A\B** (right): only the analyst-side — `export_data`, the one
  permission the analyst has that the engineer does not.

The privilege gap printout shows what an attacker gains by moving from an analyst
account to an engineer account: `delete`, `execute`, `write`. These three operations
are the crown jewels — the set-theoretic expression of why engineer accounts
require more hardening than analyst accounts.


### 2 · Input Validation as Set Membership

We model the input validation problem as set operations. The key sets are:

- $\mathcal{I}$ — all possible inputs (the full input space)
- $\mathcal{I}_{\text{valid}}$ — inputs that pass validation
- $\mathcal{I}_{\text{safe}}$ — inputs that are actually safe to process
- $\mathcal{I}_{\text{attack}}$ — inputs an attacker might send

The security goal is $\mathcal{I}_{\text{actual}} \subseteq \mathcal{I}_{\text{valid}}
\subseteq \mathcal{I}_{\text{safe}}$. The attack surface is
$\mathcal{I}_{\text{attack}} \cap \overline{\mathcal{I}_{\text{valid}}}$ —
attacker inputs that bypass validation.


```python
# ── 2 · Input Validation as Set Membership ───────────────────────────────────

# Simulated input domain: SQL fragments the system might receive
# (simplified for illustration)
all_inputs = {
    'SELECT * FROM users',
    'SELECT name FROM products WHERE id=1',
    "SELECT * FROM users WHERE name='' OR '1'='1'",  # SQL injection
    'INSERT INTO logs VALUES (1, "login")',
    "'; DROP TABLE users; --",                        # SQL injection
    'UPDATE settings SET value=1 WHERE key="debug"',
    "admin' --",                                       # SQL injection
    'SELECT count(*) FROM orders',
    "1; EXEC xp_cmdshell('whoami')",                  # command injection
    'SELECT id FROM sessions WHERE token=?',
}

# Validation rule: reject strings containing SQL meta-characters
SQL_DANGER = {"'", '"', ';', '--', 'DROP', 'EXEC', 'xp_', "OR '1'"}

def validate(s):
    return not any(danger.lower() in s.lower() for danger in SQL_DANGER)

# Compute the sets
I_valid   = {s for s in all_inputs if validate(s)}
I_invalid = all_inputs - I_valid   # complement relative to all_inputs

print(f"Total inputs:       |I|           = {len(all_inputs)}")
print(f"Valid inputs:       |I_valid|      = {len(I_valid)}")
print(f"Invalid inputs:     |I_invalid|    = {len(I_invalid)}")
print(f"\nI_invalid (attack surface):")
for s in sorted(I_invalid):
    print(f"  ✗  {s}")

print(f"\nI_valid (pass validation):")
for s in sorted(I_valid):
    print(f"  ✓  {s}")

# ── Visualise as a bar chart of categories ─────────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 3.5))

categories = ['All inputs
|I|', 'Valid
|I_valid|', 'Invalid
|I_invalid|
(attack surface)']
counts     = [len(all_inputs), len(I_valid), len(I_invalid)]
colors     = [TS_GRAY, TS_GREEN, TS_RED]

bars = ax.bar(categories, counts, color=colors, edgecolor='white',
              width=0.5, zorder=3)
for bar, n in zip(bars, counts):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
            str(n), ha='center', va='bottom', fontsize=13, fontweight='bold')

ax.set_ylabel('Number of inputs')
ax.set_title(
    'Input Validation as Set Membership
'
    r'$\mathcal{I}_{\mathrm{invalid}} = \mathcal{I} \setminus \mathcal{I}_{\mathrm{valid}}$ — the attack surface',
    pad=10
)
ax.set_ylim(0, max(counts) + 1.5)
ax.grid(axis='y', alpha=0.4)
plt.tight_layout()
plt.show()

# The formal claim and its violation
print(f"\nFormal safety claim:")
print(f"  ∀s ∈ actual_inputs, validate(s)  ≡  actual_inputs ⊆ I_valid")
print(f"\nViolation condition:")
print(f"  actual_inputs ∩ I_invalid ≠ ∅")
print(f"  (there exists an input in actual_inputs that fails validation)")
```

**What do you see?**

Three of the ten inputs fail validation — they contain SQL injection patterns.
The invalid set is the **attack surface**: every input in
$\mathcal{I}_{\text{invalid}}$ represents a path to a SQL injection vulnerability
if the system processes it.

The formal claim at the bottom connects back to Module 02 directly: the $\forall$
statement over inputs is equivalent to the subset claim
$\mathcal{I}_{\text{actual}} \subseteq \mathcal{I}_{\text{valid}}$. A violation
is a non-empty intersection of actual inputs with the invalid set. The two
framings — predicate logic and set theory — are saying the same thing in
different notation.


### 3 · Power Set and the Exponential Explosion of Permission Sets

We compute $\mathcal{P}(\text{Permissions})$ for increasing numbers of base
permissions and visualise the exponential growth. This is the formal explanation
for why role-based access control scales where permission-per-user assignment does not.


```python
# ── 3 · Power Set — Exponential Growth ───────────────────────────────────────

# Enumerate power set for small n
base_perms = ['read', 'write', 'execute', 'delete']
n = len(base_perms)

power_set = []
for r in range(n + 1):
    for combo in itertools.combinations(base_perms, r):
        power_set.append(frozenset(combo))

print(f"Base permissions: {base_perms}")
print(f"|P(Permissions)| = 2^{n} = {2**n}")
print(f"\nAll {len(power_set)} possible permission subsets:")
for i, pset in enumerate(power_set):
    label = '{' + ', '.join(sorted(pset)) + '}' if pset else '∅ (no permissions)'
    print(f"  {i:>2}. {label}")

# ── Exponential growth chart ──────────────────────────────────────────────────
ns     = list(range(1, 25))
counts = [2**n for n in ns]

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(ns, counts, color=TS_BLUE, lw=2.5, marker='o', markersize=5, zorder=3)
ax.fill_between(ns, counts, alpha=0.12, color=TS_BLUE)

# Annotate key milestones
milestones = [(4, '16
(toy system)'), (10, '1,024'), (16, '65,536'),
              (20, '~1M'), (24, '~16M')]
for x, lbl in milestones:
    y = 2**x
    ax.annotate(lbl, xy=(x, y), xytext=(x + 0.8, y * 1.8),
                fontsize=8, color=TS_AMBER,
                arrowprops=dict(arrowstyle='->', color=TS_AMBER, lw=1))

ax.set_yscale('log')
ax.set_xlabel('Number of base permissions $n$')
ax.set_ylabel('$|\mathcal{P}(\mathrm{Permissions})| = 2^n$  (log scale)')
ax.set_title(
    'Power Set Size — Why RBAC Scales and Per-User Permission Assignment Does Not
'
    'Each additional permission doubles the number of possible permission configurations',
    pad=10
)
ax.grid(True, which='both', alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nWith 30 permissions: 2^30 = {2**30:,} possible permission sets")
print(f"RBAC solution: assign users to one of k roles (k << 2^n)")
print(f"Reduces the access control problem from O(2^n) to O(k)")
```

**What do you see?**

The listing shows all 16 subsets of a four-permission set — from $\emptyset$
(no access) to $\{\text{read, write, execute, delete}\}$ (full access). Every
row is a possible permission assignment for a user.

The growth chart makes the scaling problem visceral. At $n = 10$ permissions
there are already 1,024 possible permission sets — manageable. At $n = 20$
there are over a million. At $n = 30$, more than a billion.

RBAC's solution is elegant: instead of picking from $2^n$ permission subsets,
define $k$ roles (typically $k \ll 2^n$) and assign each user to one or a few
roles. The access control problem shrinks from $O(2^n)$ configurations to $O(k)$.
The mathematics of *why* RBAC scales is the power set, and the mathematics of
what "assigning a user to a role means" is the relation — which we formalise in
Unit 02.

## Summary

| Concept | Definition | Security application |
|---|---|---|
| **Set** | Unordered collection of distinct elements | Permission sets, input domains, user groups |
| **$\in$, $\notin$** | Membership | Is this permission granted? Is this input valid? |
| **$\subseteq$** | Subset — $\forall x, x \in A \Rightarrow x \in B$ | Least privilege: actual perms ⊆ permitted perms |
| **$\cup$** | Union — set OR | Role merging; combining permission grants |
| **$\cap$** | Intersection — set AND | Mandatory access control; common permissions |
| **$\setminus$** | Difference | Privilege gap; what an attacker gains by escalation |
| **$\overline{A}$** | Complement | Invalid inputs; denied resources |
| **$\triangle$** | Symmetric difference | Policy change delta; what changed |
| **$\times$** | Cartesian product | Full access matrix space |
| **$\mathcal{P}(A)$** | Power set — $2^{|A|}$ subsets | Exponential permission configurations |
