## 🔣 Symbol Primer

Relations and functions introduce notation for structured connections between sets.

| Symbol | Name | Read it as | Meaning |
|---|---|---|---|
| $R \subseteq A \times B$ | Binary relation | "$R$ is a relation from $A$ to $B$" | $R$ is a set of ordered pairs from $A \times B$ |
| $aRb$ or $(a, b) \in R$ | Related | "$a$ is related to $b$ by $R$" | The pair $(a,b)$ is in the relation |
| $R^{-1}$ | Inverse relation | "$R$ inverse" | $\{(b,a) \mid (a,b) \in R\}$ — flip all pairs |
| $R^2$, $R \circ R$ | Composition | "$R$ composed with $R$" | $(a,c) \in R^2$ iff $\exists b, aRb \wedge bRc$ |
| $[a]_R$ | Equivalence class | "class of $a$ under $R$" | $\{b \mid aRb\}$ — all elements related to $a$ |
| $f: A \to B$ | Function | "$f$ maps $A$ to $B$" | A relation where each $a$ has exactly one image |
| $f(a)$ | Image | "$f$ of $a$" | The unique $b \in B$ that $a$ maps to |
| $\text{dom}(f)$ | Domain | "domain of $f$" | The set $A$ — all valid inputs |
| $\text{cod}(f)$ | Codomain | "codomain of $f$" | The set $B$ — all possible outputs |
| $\text{ran}(f)$ | Range / Image | "range of $f$" | $\{f(a) \mid a \in A\}$ — actual outputs |

## 1 · Binary Relations

A **binary relation** $R$ from set $A$ to set $B$ is a subset of the Cartesian
product $A \times B$:

$$R \subseteq A \times B$$

We write $aRb$ or $(a, b) \in R$ to say "$a$ is related to $b$."

### From Predicates to Relations

In Module 02 we defined two-argument predicates like $\text{can\_access}(u, r)$.
A relation is the *extension* of such a predicate — the set of all pairs for which
it is true:

$$R_{\text{access}} = \{(u, r) \in \text{Users} \times \text{Resources} \mid \text{can\_access}(u, r)\}$$

This is the **access control matrix** expressed as a binary relation.

### Representing Relations

A relation on a finite set can be represented three ways:

**As a set of pairs:**
$$R = \{(\text{Alice}, \text{payroll}),\, (\text{Alice}, \text{logs}),\, (\text{Bob}, \text{logs})\}$$

**As a Boolean matrix** (rows = $A$, columns = $B$, 1 = related):

| | payroll | source | logs |
|---|:---:|:---:|:---:|
| Alice | 1 | 0 | 1 |
| Bob | 0 | 0 | 1 |
| Carol | 0 | 1 | 1 |

**As a directed graph** (nodes = elements, arrows = related pairs) — this is the
representation we will develop fully in Module 04.


## 2 · Properties of Relations on a Single Set

When $R \subseteq A \times A$ (a relation *on* a set, from $A$ to itself), four
structural properties determine how it behaves.

### Reflexive

$$\forall a \in A,\; (a, a) \in R$$

*Every element is related to itself.* Every node has a self-loop in the directed graph.

**Security:** The "same security level" relation in mandatory access control is
reflexive — every object is at the same level as itself.

### Symmetric

$$\forall a, b \in A,\; (a, b) \in R \Rightarrow (b, a) \in R$$

*If $a$ relates to $b$, then $b$ relates to $a$.* All arrows come in pairs.

**Security:** "Is a peer of" between employees — symmetric. "Has read access to"
between users and resources — typically *not* symmetric.

### Antisymmetric

$$\forall a, b \in A,\; (a, b) \in R \wedge (b, a) \in R \Rightarrow a = b$$

*The only way for both $(a,b)$ and $(b,a)$ to hold is if $a = b$.*
Arrows never come in opposing pairs except self-loops.

**Security:** Permission level ordering — if level $A$ ≥ level $B$ and level $B$
≥ level $A$, then the levels are equal. Antisymmetry is the property of a
**partial order**.

### Transitive

$$\forall a, b, c \in A,\; (a, b) \in R \wedge (b, c) \in R \Rightarrow (a, c) \in R$$

*If $a$ relates to $b$ and $b$ relates to $c$, then $a$ relates to $c$.*

**Security:** Role hierarchy inheritance — if role $A$ inherits from role $B$
and $B$ inherits from $C$, then $A$ inherits from $C$ (we proved this in Module 00
as a direct proof of transitivity of $\subseteq$).

### Equivalence Relations

A relation that is **reflexive, symmetric, and transitive** is an
**equivalence relation**. It partitions the set $A$ into disjoint
**equivalence classes** where all related elements form one class.

**Security:** "Runs at the same trust level" is an equivalence relation over
processes — it partitions all running processes into trust level groups.


## 3 · Functions

A **function** $f: A \to B$ is a binary relation from $A$ to $B$ with one
additional constraint: every element of $A$ maps to **exactly one** element of $B$.

$$f \subseteq A \times B \quad\text{such that}\quad
\forall a \in A,\; \exists!\, b \in B,\; (a, b) \in f$$

*(recall $\exists!$ from Module 02 — there exists a unique)*

### Total vs. Partial Functions

A **total function** is defined for every element of $A$ — every input has an output.
A **partial function** is defined only for *some* elements of $A$ — some inputs
are undefined (like division by zero).

**Security:** An access control decision function must be **total** — every
access request must produce a decision (allow or deny). A partial access function
that is undefined for some requests is a security gap.

### Injective, Surjective, Bijective

These three properties classify how a function maps its domain to its codomain.
They are the foundation of Unit 03.

**Injective** (one-to-one) — $f: A \to B$ is injective if distinct inputs always
produce distinct outputs:

$$\forall a_1, a_2 \in A,\; f(a_1) = f(a_2) \Rightarrow a_1 = a_2$$

Equivalently: no two inputs share the same output. In the directed graph, no
two arrows point to the same target.

**Security:** A hash function used for deduplication must be injective (collision-free)
— two different inputs must not produce the same hash. Collision attacks break injectivity.

**Surjective** (onto) — $f: A \to B$ is surjective if every element of $B$
is the image of at least one element of $A$:

$$\forall b \in B,\; \exists a \in A,\; f(a) = b$$

The range equals the codomain: $\text{ran}(f) = B$.

**Security:** A pseudorandom number generator is expected to be surjective over
its output range — every possible output value should be reachable.

**Bijective** (one-to-one and onto) — both injective and surjective.
Every input has a unique output and every output has a unique input.
Bijections are **invertible** — $f^{-1}$ exists and is also a function.

**Security:** Encryption functions must be bijections — every plaintext maps
to a unique ciphertext (injective) and every possible ciphertext corresponds
to some plaintext (surjective). We formalise this in Unit 03.


## 4 · Equivalence Classes and Security Partitions

Given an equivalence relation $R$ on $A$, the **equivalence class** of element
$a$ is:

$$[a]_R = \{b \in A \mid aRb\}$$

The collection of all equivalence classes forms a **partition** of $A$ — a set
of non-empty, pairwise disjoint subsets whose union is all of $A$.

$$A = [a_1]_R \cup [a_2]_R \cup \cdots \cup [a_k]_R \quad\text{where each pair is disjoint}$$

### Security Partitions in Practice

**Mandatory Access Control (MAC) levels.** "Runs at the same classification level"
is an equivalence relation on processes and data objects. The equivalence classes
are the classification levels: UNCLASSIFIED, CONFIDENTIAL, SECRET, TOP SECRET.
The Bell-LaPadula model's "read down, write up" rules are defined in terms of
this partition.

**Network segmentation.** "Is in the same network segment" is an equivalence relation
on hosts. The equivalence classes are the segments. Zero-trust architecture
challenges this partition — it argues that the equivalence classes should each
contain exactly one host (every host in its own segment).

**User groups.** "Has the same base permissions" is an equivalence relation on users.
The equivalence classes are the roles. RBAC is the formal consequence of
organising users by permission equivalence classes.

### 1 · Access Control Matrix as a Binary Relation

We model the access control policy for a small system as a binary relation
$R_{\text{access}} \subseteq \text{Users} \times \text{Resources}$.
Then we query it formally and compute the inverse relation $R^{-1}$ —
which flips the question from "what can this user access?" to "who can access
this resource?"


```python
# ── 1 · Access Control Relation ──────────────────────────────────────────────

users     = ['Alice', 'Bob', 'Carol', 'Dave', 'Eve']
resources = ['payroll', 'source_code', 'logs', 'config', 'public_docs']

# R_access as a set of (user, resource) pairs
R_access = {
    ('Alice', 'payroll'), ('Alice', 'source_code'),
    ('Alice', 'logs'),    ('Alice', 'config'),     ('Alice', 'public_docs'),
    ('Bob',   'source_code'), ('Bob', 'logs'),     ('Bob',  'public_docs'),
    ('Carol', 'payroll'), ('Carol', 'logs'),        ('Carol','public_docs'),
    ('Dave',  'public_docs'),
    ('Eve',   'logs'),    ('Eve',   'source_code'), ('Eve',  'public_docs'),
}

# ── Heatmap ────────────────────────────────────────────────────────────────────
matrix = np.zeros((len(users), len(resources)), dtype=float)
for u_idx, user in enumerate(users):
    for r_idx, res in enumerate(resources):
        if (user, res) in R_access:
            matrix[u_idx, r_idx] = 1.0

fig, axes = plt.subplots(1, 2, figsize=(13, 4.5))
cmap = ListedColormap([TS_RED, TS_GREEN])

# Original relation
ax = axes[0]
ax.imshow(matrix, cmap=cmap, vmin=0, vmax=1, aspect='auto')
for r in range(len(users)):
    for c in range(len(resources)):
        ax.text(c, r, '✓' if matrix[r,c] else '✗',
                ha='center', va='center', fontsize=12,
                color='white', fontweight='bold')
ax.set_xticks(range(len(resources)))
ax.set_xticklabels(resources, rotation=30, ha='right', fontsize=9)
ax.set_yticks(range(len(users)))
ax.set_yticklabels(users, fontsize=10)
ax.set_title('$R_{\\mathrm{access}}$ — who can access what', fontweight='bold', pad=8)

# Inverse relation R^{-1}: resources × users
ax2 = axes[1]
ax2.imshow(matrix.T, cmap=cmap, vmin=0, vmax=1, aspect='auto')
for r in range(len(resources)):
    for c in range(len(users)):
        ax2.text(c, r, '✓' if matrix[c,r] else '✗',
                 ha='center', va='center', fontsize=12,
                 color='white', fontweight='bold')
ax2.set_xticks(range(len(users)))
ax2.set_xticklabels(users, rotation=30, ha='right', fontsize=9)
ax2.set_yticks(range(len(resources)))
ax2.set_yticklabels(resources, fontsize=9)
ax2.set_title('$R^{-1}_{\\mathrm{access}}$ — who has access to each resource',
              fontweight='bold', pad=8)

plt.suptitle('Access Control Relation and Its Inverse', fontsize=12,
             fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# ── Formal queries ────────────────────────────────────────────────────────────
def accessible_by(user):
    return {r for (u, r) in R_access if u == user}

def accessed_by(resource):
    return {u for (u, r) in R_access if r == resource}

print("Formal relation queries:")
for user in users:
    print(f"  {user:<8} can access: {sorted(accessible_by(user))}")
print()
print("  Resource          accessible by:")
for res in resources:
    print(f"  {res:<15}  {sorted(accessed_by(res))}")

# ── Security finding: resources with only ONE authorised user ──────────────────
single_access = [(r, accessed_by(r)) for r in resources if len(accessed_by(r)) == 1]
if single_access:
    print(f"\n⚠  Single-point-of-access resources (availability risk):")
    for res, users_with_access in single_access:
        print(f"   {res}: only {list(users_with_access)[0]} has access")
```

**What do you see?**

The left heatmap is the standard access control matrix — rows are users, columns
are resources, green means permitted. The right heatmap is the **inverse relation**
$R^{-1}$ — rows are now resources, showing which users can access each one.

The inverse relation answers the auditor's question: *"If this resource is
compromised, which user accounts are implicated?"* — the formal set-theoretic
version of blast-radius analysis.

The availability-risk finding at the bottom demonstrates another use: `config`
is accessible by only one user (Alice). If Alice's account is disabled, no one
can access the configuration. The binary relation makes this structural weakness
formally visible.


### 2 · Relation Property Checker

The four properties — reflexive, symmetric, antisymmetric, transitive — are each
a $\forall$ statement. We implement each as a predicate check and classify several
security-relevant relations automatically, then display the results as a directed
graph to make the structure visual.


```python
# ── 2 · Relation Property Checker ────────────────────────────────────────────

def is_reflexive(A, R):
    return all((a, a) in R for a in A)

def is_symmetric(A, R):
    return all((b, a) in R for (a, b) in R)

def is_antisymmetric(A, R):
    return all(a == b for (a, b) in R if (b, a) in R)

def is_transitive(A, R):
    return all((a, c) in R
               for (a, b) in R
               for (b2, c) in R
               if b == b2)

def classify_relation(name, A, R):
    props = {
        'Reflexive':      is_reflexive(A, R),
        'Symmetric':      is_symmetric(A, R),
        'Antisymmetric':  is_antisymmetric(A, R),
        'Transitive':     is_transitive(A, R),
    }
    is_equiv = props['Reflexive'] and props['Symmetric'] and props['Transitive']
    is_order = props['Reflexive'] and props['Antisymmetric'] and props['Transitive']
    kind = 'Equivalence relation' if is_equiv else ('Partial order' if is_order else 'General relation')
    return props, kind

# ── Define relations ──────────────────────────────────────────────────────────
levels = ['L0', 'L1', 'L2', 'L3']   # security clearance levels 0–3

# "≤" on clearance levels — partial order
R_order = {(a, b) for a in levels for b in levels
           if int(a[1]) <= int(b[1])}

# "same level" on clearance levels — equivalence
R_equiv = {(a, b) for a in levels for b in levels
           if a == b}

# "can communicate with" in a simplified network (symmetric)
hosts  = ['h1', 'h2', 'h3', 'h4']
R_comm = {('h1','h2'), ('h2','h1'), ('h2','h3'), ('h3','h2'),
          ('h1','h1'), ('h2','h2'), ('h3','h3'), ('h4','h4'),
          ('h1','h3'), ('h3','h1')}  # h4 is isolated

# "has more permissions than" — NOT transitive (to show a failure case)
perms = ['guest', 'user', 'mod', 'admin']
R_priv = {('user','guest'), ('mod','guest'), ('mod','user'), ('admin','guest'),
          ('admin','user'), ('admin','mod')}  # strict partial order (no reflexive)

relations = [
    ('Clearance level ≤  (partial order)', levels, R_order),
    ('Same clearance level  (equivalence)', levels, R_equiv),
    ('Host communication graph', hosts, R_comm),
    ('Strict privilege dominance', perms, R_priv),
]

# ── Display results ────────────────────────────────────────────────────────────
print(f"{'Relation':<40} {'Refl':^6} {'Symm':^6} {'Anti':^6} {'Trans':^6}  Classification")
print("─" * 85)
for name, A, R in relations:
    props, kind = classify_relation(name, A, R)
    marks = [('✓' if v else '✗') for v in props.values()]
    print(f"  {name:<38} " + "  ".join(f"{m:^6}" for m in marks) + f"  {kind}")

# ── Draw directed graph for the clearance partial order ───────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

for ax, (name, A, R), pos_fn in [
    (axes[0], relations[0],  # partial order
     lambda: {'L0':(0,0),'L1':(1,1),'L2':(2,2),'L3':(3,3)}),
    (axes[1], relations[1],  # equivalence
     lambda: {'L0':(0,1),'L1':(1,2),'L2':(2,1),'L3':(3,0)}),
]:
    G = nx.DiGraph()
    G.add_nodes_from(A)
    G.add_edges_from(R)
    pos = pos_fn()

    self_loops = [(u,v) for u,v in G.edges() if u == v]
    other_edges = [(u,v) for u,v in G.edges() if u != v]

    nx.draw_networkx_nodes(G, pos, ax=ax, node_color=TS_BLUE,
                           node_size=800, alpha=0.9)
    nx.draw_networkx_labels(G, pos, ax=ax, font_color='white',
                            font_size=11, font_weight='bold')
    nx.draw_networkx_edges(G, pos, edgelist=other_edges, ax=ax,
                           edge_color=TS_AMBER, arrows=True,
                           arrowsize=20, width=2,
                           connectionstyle='arc3,rad=0.1')
    if self_loops:
        nx.draw_networkx_edges(G, pos, edgelist=self_loops, ax=ax,
                               edge_color=TS_GREEN, arrows=True,
                               arrowsize=15, width=1.5)

    props, kind = classify_relation(name, A, R)
    ax.set_title(f'{name}
({kind})', fontsize=9, fontweight='bold', pad=8)
    ax.axis('off')
    ax.set_facecolor('white')

plt.suptitle('Binary Relations as Directed Graphs', fontsize=12,
             fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

**What do you see?**

The property table classifies all four relations instantly:

- **Clearance ≤** is a partial order (reflexive, antisymmetric, transitive) —
  the expected structure for a privilege hierarchy.
- **Same clearance** is an equivalence relation (reflexive, symmetric, transitive) —
  it partitions levels into classes of equal standing.
- **Communication graph** is symmetric but not transitive — just because h1 talks
  to h2 and h2 talks to h3 does not mean h1 talks to h3 (a firewall rule blocks it).
  This is what network segmentation *is* at the formal level: deliberately breaking
  transitivity.
- **Strict privilege dominance** has no reflexive pairs (admin does not dominate
  itself) — it is a strict partial order.

The directed graph makes the structural difference visible: the partial order
has a clear "uphill" direction with self-loops (amber arrows up, green self-loops);
the equivalence relation has self-loops only (every node relates only to itself in
this case where each level is its own class).


### 3 · Equivalence Classes — Trust Level Partitions

We compute the equivalence classes of a "runs at the same trust level" relation
on a set of processes and visualise the resulting partition. Each class becomes
a security zone.


```python
# ── 3 · Equivalence Classes — Process Trust Partitions ───────────────────────

processes = {
    'kernel':      0,   # trust level 0 = highest
    'systemd':     0,
    'sshd':        1,
    'nginx':       1,
    'app_server':  2,
    'worker_1':    2,
    'worker_2':    2,
    'logger':      3,
    'monitor':     3,
    'user_shell':  4,   # trust level 4 = lowest
    'browser':     4,
    'untrusted':   4,
}

# "Same trust level" equivalence relation
R_trust = {(p1, p2)
           for p1, level1 in processes.items()
           for p2, level2 in processes.items()
           if level1 == level2}

# Compute equivalence classes
def equivalence_classes(A_dict, R):
    visited = set()
    classes = []
    for a in A_dict:
        if a not in visited:
            cls = frozenset(b for b in A_dict if (a, b) in R)
            classes.append(cls)
            visited |= cls
    return sorted(classes, key=lambda c: min(A_dict[x] for x in c))

classes = equivalence_classes(processes, R_trust)
level_names = {0: 'Kernel (L0)', 1: 'System (L1)',
               2: 'Application (L2)', 3: 'Monitoring (L3)', 4: 'User (L4)'}

print("Equivalence classes of 'same trust level':")
print(f"  (Partition of {len(processes)} processes into {len(classes)} classes)
")
for cls in classes:
    level = processes[next(iter(cls))]
    print(f"  [{level_names[level]}]  {sorted(cls)}")

# ── Visualise partition as grouped nodes ──────────────────────────────────────
fig, ax = plt.subplots(figsize=(12, 5.5))
ax.set_xlim(0, 13)
ax.set_ylim(-0.5, len(classes) + 0.5)
ax.axis('off')
ax.set_facecolor('white')
fig.patch.set_facecolor('white')

zone_colors = [TS_RED, TS_AMBER, TS_BLUE, TS_GREEN, TS_GRAY]
zone_labels = list(level_names.values())

for i, cls in enumerate(classes):
    level  = processes[next(iter(cls))]
    color  = zone_colors[level]
    procs  = sorted(cls)
    y      = len(classes) - 1 - i

    # Zone label
    ax.text(0.1, y, zone_labels[level], va='center', fontsize=10,
            fontweight='bold', color=color)

    # Process boxes
    for j, proc in enumerate(procs):
        x = 2.2 + j * 1.7
        rect = mpatches.FancyBboxPatch(
            (x - 0.7, y - 0.32), 1.4, 0.64,
            boxstyle='round,pad=0.08',
            facecolor=color, edgecolor='white', linewidth=1.5, alpha=0.85
        )
        ax.add_patch(rect)
        ax.text(x, y, proc, ha='center', va='center',
                fontsize=8, color='white', fontweight='bold')

    # Horizontal zone separator
    if i < len(classes) - 1:
        ax.axhline(y - 0.5, color='#cccccc', lw=1, alpha=0.7)

ax.set_title(
    'Equivalence Classes of "Same Trust Level"
'
    'Each class = one security zone   |   Partition is disjoint and exhaustive',
    fontsize=11, fontweight='bold', pad=10
)
plt.tight_layout()
plt.show()

# Verify partition properties
all_procs = set(processes.keys())
union_of_classes = set().union(*classes)
pairwise_disjoint = all(
    cls_i & cls_j == frozenset()
    for cls_i, cls_j in itertools.combinations(classes, 2)
)
print(f"\nPartition verification:")
print(f"  Union of all classes = all processes: {union_of_classes == all_procs}")
print(f"  All classes pairwise disjoint:        {pairwise_disjoint}")
print(f"  → Valid partition: {union_of_classes == all_procs and pairwise_disjoint} ✓")
```

**What do you see?**

The twelve processes are partitioned into five equivalence classes — one per
trust level. The partition verification at the bottom confirms two things:

1. **Exhaustive:** every process appears in exactly one class (union = all processes).
2. **Disjoint:** no process appears in two classes (pairwise intersection = $\emptyset$).

These are precisely the properties that make it a valid partition, and they follow
directly from the equivalence relation being reflexive, symmetric, and transitive.

**The security implication.** The equivalence classes are the security zones.
Zero-trust architecture says these zones should each contain exactly one host —
every process in its own equivalence class, no implicit trust between peers.
The mathematics of "what a security zone *is*" is the equivalence class.

## Summary

| Concept | Definition | Security meaning |
|---|---|---|
| **Binary relation** | $R \subseteq A \times B$ | Access control matrix, any two-object policy |
| **Reflexive** | $\forall a, aRa$ | Self-access; same-level as self |
| **Symmetric** | $aRb \Rightarrow bRa$ | Peer relationship; bidirectional trust |
| **Antisymmetric** | $aRb \wedge bRa \Rightarrow a=b$ | Hierarchical ordering |
| **Transitive** | $aRb \wedge bRc \Rightarrow aRc$ | Inherited access; role chaining |
| **Equivalence relation** | Reflexive + Symmetric + Transitive | Trust level; security zone membership |
| **Equivalence classes** | Disjoint, exhaustive partition | Security zones; RBAC roles |
| **Function** | Each input has exactly one output | Total access decision; deterministic policy |
| **Injective** | Distinct inputs → distinct outputs | Collision-free hash; unique session tokens |
| **Surjective** | Every output reached | PRNG coverage; complete codomain |
| **Bijective** | Injective + Surjective = invertible | Encryption / decryption (Unit 03) |
